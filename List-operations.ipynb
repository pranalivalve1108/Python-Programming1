```python
# Numbers list
num = [10, 20, 30, 40, 50]
print(num)
```

```python
# Letters list
letter = ['a', 'b', 'c', 'd']
print(letter)
```

```python
# Characters List
char = ["How", "are", "you", "Pranali"]
print(char)
```

```python
# Mix list
mix = [10, 'a', 20, "Swara", 'b', 10.5]
print(mix)
```

```python
# Nested list 
matrix = [[10, 'a', 20], [30, 'b', "Pranali"], ['x', "Swara", 50]]
print(matrix)
```

```python
# Accessing list elements
print(num[1])
```

```python
#Updating list elements
mix[3] = "Neha"
print(mix)
```

```python
print(mix[-2]) #print 2nd location from end
```

```python
print(mix[2:])
```

```python
print(mix[:2])
```

```python
print(mix[1:4])
```

```python
print(mix[::2]) #skip 1 value 
```

```python
print(mix[::3]) #skip 2 value
```

```python
#list in reverse order
print(mix[::-1])
```

```python
#List operations
#1. Add element
fruits = ["Apple", "Banana", "Cherry"]
print(fruits)
fruits.append("Orange")
print(fruits)
fruits.insert(1, "Strawberry")
print(fruits)
```

```python
#2.Remove element
fruits.remove("Banana")
print(fruits)
```

```python
fruits.pop(0)
print(fruits)
```

```python
fruits.clear()
```

```python
print(fruits)
```
