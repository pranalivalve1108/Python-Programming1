```python
# Arithmetic Operators
a = 10
b = 20
print("Addition is:", a + b)
print("Sutraction is:", a - b)
print("Multiplication is:", a * b)
print("Division is:", a / b)
print("Modulus is:", a % b)    #remainder
print("Exponent(Power) of is:", a ** b)
print("Floor Division is:", a // b)
```

```python
#Comparison operators
a = 10
b = 20

print(a == b)   #Equal to
print(a != b)   #Not equal to
print(a > b)   #Greater than
print(a < b)   #Less than
print(a <= b)   #less than equal to
print(a >= b)   #Greater than Equal to
```

```python
#Logical operators
#and operator
age = 20
has_id = True
if age > 10 and has_id:
    print("Eligible to vote")
    
```

```python
#or operator
weekend = False
holiday = True
if weekend or holiday:
    print("You can relax today")
```

```python
#not operator
is_raining = False
if not is_raining:
    print("Go for a walk")
```

```python
#Conditional Statements
#IF 
a = 10
if a < 20:
    print("A is Less than 20")
```

```python
#If - else
a = 20
if a < 10:
    print("A is less than 10")
else:
    print("A is Greate than 10")
```

```python
#If - elif -else
marks = 80
if marks < 90:
    print("Distinction")
elif marks < 70:
    print("First class")
else:
    print("Fail")
```

```python
#Loops in Python
#for Loop
a= [10, 20, 30, 40, 50]
for i in a:
    print(i)
```

```python
#using range()
for i in range(10):
    print(i)
```

```python
#While loop
count = 1
while count <= 5:
    print(count)
    count+=1
```

```python
#Loop control statements
#1. Break
for i in range(10):
    if i == 5:
        break;
    print(i)
```

```python
#2. Continue
for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
#3. Pass
for i in range(5):
    pass
```

```python
#Functions in python
#Define a function
def greet():
    print("Greetings of the day...!!!")
#Call the function
greet()
```

```python
#Functions with arguments
def add(a, b):
    print("Addition of A & B is:", a+b)

add(10, 20)
```

```python
#Function with return values
def multiply(x, y):
    return x*y
multiply(10, 5)
```

```python
#defalut arguments
def welcome(name="guest"):
    print("Hello", name)
welcome()
welcome("Pranali")
```

```python
#Keyword arguments
def student(name, age):
    print(name, age)
student(age=21, name="Pranali")
```

```python
#Variable length arguments
#1. *args
def sum_all(*numbers):
    return sum(numbers)
print(sum_all(1, 2, 3, 4, 5))
```

```python
#2. **kwargs
def details(**info):
    print(info)
details(name="Pranali", age=20)
```

```python
#Lambda function
square = lambda x:x*x
print(square(5))
```

```python
#Built in functions
a = [1, 2, 3, 4, 5]
print("Length is:",len(a))
print("Sum is:", sum(a))
print("Maximum number is:", max(a))
print("Minimum number is:", min(a))
print("Type of A is:", type(a))
```
